# 3 Example

At https://github.com/sawula/Survey-of-Python-data-science-stack is the file named `PandasJumble0.csv`.  

* Read the data into a dataframe.  
* Select and sort so that the name `NELLIE` appears in a column (in that order)
* Repeat for `TOMA`.

## **SOLUTION**

First, follow the link to the GitHub repo.  Click on the file.  Click on `Raw`.  Copy the URL.  Etc.

```python
jumble0 = pd.read_csv('https://raw.githubusercontent.com/sawula/Survey-of-Python-data-science-stack/gh-pages/PandasJumble0.csv')
jumble0
```

I'm guessing that the letters for `NELLIE` have `Catg` of `k` (and that those for `TOMA` have `Catg` `h`).  I'm going to select those out.

```python
jumble0_k = jumble0[jumble0.Catg=='k']
jumble0_k
```

Now I'm going to sort these according to the `Valu` column.

```python
jumble0_k.sort_values('Valu')
```

Winner!  

Now let's quickly find `TOMA`.

```python
jumble0_h = jumble0[jumble0.Catg=='h']
jumble0_h
```

```python
jumble0_h.sort_values('Valu')
```

Oops.  These are upside down.  We'll have to sort in descending order.

```python
jumble0_h.sort_values('Valu',ascending=False)
```

Winner!
